# Tipos de Análise de Dados

* Descritiva
* Exploratória
* Inferencial
* Preditiva
* Causal
* Mecanicista

## Análise Descritiva

Objetivo: descrever um conjunto de dados

* Média: $$\overline{x} = \frac{\sum_{x_i}}{n}$$

* Moda: mais frequente

* Mediana: divide os dados ordenados em dois

* Variância e Desvio Padrão

### Uma ideia de como os dados se distribuem e seus possíveis outliers?

* Boxplot



### Apenas sumarização dos dados é suficiente?
* Quarteto de Anscombe



* estatísticas descritivas simples são quase idênticas nos quatro datasets



* a representação gráfica é completamente diferente



## Análise Exploratória

Objetivo: Encontrar relações/padrões antes desconhecidos

* analisar entendimento prévio
* abrir novos caminhos ou estratégias
* gráficos
* análise univariada e multivariada (relações entre variáveis)

__Isoladamente não deve ser utilizada para generalizar ou prever__

## Análise Inferencial

Objetivo: utilizar uma amostra para generalizar ou dizer algo sobre uma população.

* Testes de Hipótese
* Intervalos de Confiança
* Alguns modelos estatísticos (regressão linear,  logística, ...)

Ex: pesquisas eleitorais para saber o resultado de uma eleição com uma determinada margem de erro

## Análise Preditiva

Objetivo:  Usar algumas características de algum objeto para prever outras.
* Machine learning


## Análise Causal

Objetivo: o que acontece com uma variável quando você altera outra
* efeito médio pode não ser aplicado a um indivíduo
* Exemplo: Teste A/B. Aumento de cliques em propagandas após uma mudança de layout no site.

## Análise Mecanicista

Objetivo: Entender as exatas mudanças em variáveis que levam a mudanças em outras. Normalmente conjunto determinístico de equações. (Engenharia, Física).

* Exemplo: Simuladores; Análise de pavimentos em rodovias.

# Análise de Dados na Prática

* Tipos de dados estão corretos?
* Existem valores faltantes? Nulos? Vazios?
* Dados estão padronizados? Limpos (Masculino, Masc, F, Feminino)
* Valores são coerentes para os campos? (Idade -20, 190)
* Existem outliers? Como devo tratar?
* Como extraio valores de textos livres?
* O dataset está em formato adequado?
* Posso complementar meu dataset com informções extraídas dele mesmo? (Engenharia de features)


# O que é ciência de dados?

Ciência de dados é o uso conjunto de técnicas e ferramentas de algumas áreas científicas como Estatística e Ciência da Computação com o objetivo de analisar dados e extrair conhecimento ou insights. Ou seja, não é uma ciência e não é uma nova área do conhecimento, é um processo que começa desde a aquisição dos dados até a entrega de um produto que pode subsidiar decisões.
